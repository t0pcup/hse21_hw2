## Добавляем координаты генов

````python
lst = list()
with open('gms2.lst') as f:
    lst = f.read().splitlines()
````

````python
lst[123].split()
````

````python
lst = lst[10:]
````

````python
from pprint import pprint
from Bio.SeqFeature import SeqFeature, FeatureLocation
````

````python
cnt = -1
sc = list()
for s in lst:
  # finding SequenceID
  if not('+' in s) and not('-' in s):
    if 'SequenceID:' in s:
      print(1)
      # separating new scaffold from previous one
      if(cnt > -1):
        print(sc)
        records[cnt].features = sc
      cnt = cnt + 1
      sc = list()
    else:
      print('nothing to do')
  elif not('sequence-region' in s):
    # countinuing current scaffold
    s_split = list()
    for s_split_el in s.split():
      s_split_el = s_split_el.replace('>', '').replace('<', '')
      s_split.append(s_split_el)
    strand_el = 1 if s_split[1] == '-' else -1
    el = SeqFeature(FeatureLocation(int(s_split[2]), int(s_split[3]), strand=strand_el), type="CDS")
    el.qualifiers['locus_tag'] = [s_split[0]]
    sc.append(el)
````

## Добавляем функции белков (из БД SwissProt)

Читаем информацию из файла scaffolds.hits_from_SwissProt.txt

````python
!head scaffolds.hits_from_SwissProt.txt
````

````python
records[17].features[0].qualifiers
````

````python
lst = list()
with open('scaffolds.hits_from_SwissProt.txt') as f:
    lst = f.read().splitlines()
print(lst)
````

````python
q = "product_SwissProt"
for r in records:
  for f in r.features:
    f.qualifiers[q] = ""
    for el in lst:
      if f.qualifiers['locus_tag'][0] == el.split('\t')[0]:
        a, b = el.split('|')[2].split('\t')[0], el.split('\t')[-2]
        f.qualifiers[q] += f"SwissProt:  {a} (E_value= {b}),  "
````

````python
for r in records:
  for f in r.features:
    if f.qualifiers[q] != "":
      print(f.qualifiers[q])
````

````python
for r in records:
  for f in r.features:
    if f.qualifiers[q] != "":
      print('----------------' + f.qualifiers['locus_tag'][0] +  '----------------')
      for word in f.qualifiers[q].split(' '):
        if ('_' in word) and ('E_value' not in word):
          print(word)
````

````python
lst_prots, lst_functions = list(), list()
file1 = open("uniprot_sprot.dat", "r")

for i in range(72271384):
  if i % 1000000 == 0:
    print(i)
  line = file1.readline()
  if ('ID' in line) and ('Reviewed' in line):
    lst_prots.append(line.split()[1])
  string_function = ""
  if('FUNCTION: ' in line):
    string_function += line
    line = file1.readline()
    while(('-!-' not in line) and ('---' not in line)):
      string_function += line
      line = file1.readline()
    lst_functions.append(f"{lst_prots[-1]}  {string_function}")

file1.close
````

````python
print(len(lst_prots), len(lst_functions))
````

````python
for r in records: # finding proteins functions
  for f in r.features:
    if f.qualifiers[q] != "":
      f.qualifiers['function_SwissProt'] = ""
      for func in lst_functions:
        if func.split()[0] in f.qualifiers[q].split(' '):
          f.qualifiers['function_SwissProt'] += func
````
